# Ход выполнения работы
### 1. Создание набора данных

На первом шаге создаётся искусственный набор данных, состоящий из трёх кластеров, формируемых на основе нормального распределения и разнесённых в пространстве с помощью матрицы смещений. Каждый кластер получает своё положение, затем данные масштабируются по признакам, объединяются в одну таблицу и перемешиваются. Этот шаг важен для получения контролируемого набора данных, на котором можно корректно демонстрировать работу методов кластеризации и дальнейшей визуализации.

```python
import numpy as np
import pandas as pd

def make_data(n, seed):
    np.random.seed(seed)
    shift_matrix = np.array([[3, 3],
        [6, 9],
        [9, 3]])
    data = np.random.randn(3, 2, n) + shift_matrix.reshape((3, 2, 1))
    data = np.swapaxes(data, 1, 2)
    data = data.reshape((-1, 2))
    data *= np.array([[20, 0.5]])
    df = pd.DataFrame({'x': data[:, 0], 'y': data[:, 1]},
                      columns=['x', 'y'])
    df = df.sample(frac=1.0)
    return df
```

На этом этапе вызывается функция генерации данных, чтобы получить рабочий DataFrame с двумя признаками — x и y. Выбор параметра n определяет, сколько объектов будет создано в каждом из трёх кластеров, а параметр seed обеспечивает воспроизводимость результатов. После выполнения шага у нас появляется реальный набор данных, на котором затем будут проводиться стандартизация, кластеризация и визуализация.

```python
df = make_data(n=60, seed=42)

# Просмотр первых строк
df.head()
```

### 2. Стандартизация данных

На шаге стандартизации каждый признак приводится к общему масштабу: со средним 0 и стандартным отклонением 1. Это необходимо, потому что исходные признаки имеют разный масштаб (x крупный, y маленький), а алгоритмы кластеризации, включая KMeans, чувствительны к различиям в масштабе. После стандартизации оба признака начинают одинаково влиять на расстояния между объектами, что делает кластеризацию корректной.

```python
    from sklearn.preprocessing import StandardScaler
    
    # Стандартизация данных
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[['x', 'y']])
    
    # Преобразуем обратно в DataFrame для удобства
    df_scaled = pd.DataFrame(scaled_data, columns=['x_scaled', 'y_scaled'])
    df_scaled.head()
```

### 3. Обучение KMeans и получение меток для 3 кластеров

На этом этапе применяется алгоритм KMeans для разбиения стандартизированных данных на три кластера. Метод fit_predict одновременно обучает модель и присваивает каждому объекту номер кластера. Полученные метки показывают, к какой группе модель отнесла каждую точку. Эти данные понадобятся для визуализации и анализа качества кластеризации.

```python
from sklearn.cluster import KMeans

# Обучение модели KMeans на стандартизированных данных
kmeans_3 = KMeans(n_clusters=3, random_state=42)
labels_3 = kmeans_3.fit_predict(df_scaled[['x_scaled', 'y_scaled']])

# Добавим метки в таблицу
df_scaled['cluster_3'] = labels_3

df_scaled.head()

print("Метки кластеров (для каждого объекта):")
print(labels_3)
```

### 4. Визуализация 

На этом шаге мы визуально отображаем результат работы алгоритма KMeans. Каждая точка окрашена в цвет своего кластера, что позволяет рассмотреть компактность групп и их разделение. Центры кластеров отмечены крупными красными маркерами, что помогает понять, куда алгоритм «поставил» средние значения каждой группы. Такая визуализация важна, чтобы оценить корректность разбиения и убедиться, что модель выделила естественные структуры в данных.

```python
import matplotlib.pyplot as plt

# Координаты центров кластеров
centers = kmeans_3.cluster_centers_

plt.figure(figsize=(8, 6))

# Отрисовка точек по кластерным меткам
plt.scatter(df_scaled['x_scaled'], df_scaled['y_scaled'],
            c=df_scaled['cluster_3'], cmap='viridis', s=40)

# Отрисовка центров кластеров
plt.scatter(centers[:, 0], centers[:, 1],
            c='red', s=200, marker='X', label='Центры кластеров')

plt.title("Кластеры (n_clusters = 3)")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

На визуализации чётко выделяются три отдельные группы, каждая из которых образует плотный и компактный кластер. Верхний кластер расположен значительно выше двух остальных по оси y, а два нижних кластера хорошо разделены по оси x. Центры, отмеченные красными маркерами, находятся внутри соответствующих групп и отражают их среднее положение. Разделение между кластерами выражено явно, пересечений практически нет, что подтверждает корректность выбора количества кластеров и успешную работу алгоритма KMeans для данного набора данных.

```python
from sklearn.metrics import silhouette_score

# Коэффициент силуэта
sil_score_3 = silhouette_score(df_scaled[['x_scaled', 'y_scaled']], labels_3)
print("Качество кластеризации по метрике силуэта:", sil_score_3)

# Добавим кластеры в исходный df
df['cluster_3'] = labels_3

# Средние значения признаков по кластерам (центры до стандартизации)
cluster_means_3 = df.groupby('cluster_3').mean()
print(cluster_means_3)
```

По средним значениям признаков видно, что алгоритм KMeans чётко разделил выборку на три естественные группы. Кластер 1 содержит объекты с наименьшими значениями признака x (около 58) и низкими значениями y — это левая нижняя группа на графике. Кластер 2 смещён вправо и характеризуется самым высоким значением x (около 181) при тех же низких значениях y — это правая нижняя группа. Кластер 0 выделяется по вертикальному признаку: среднее значение y около 4.53 заметно выше остальных при среднем значении x. Этот кластер соответствует верхней группе точек. Такое разбиение полностью совпадает со структурой данных: два кластера различаются по оси x, а третий — по оси y. Значение коэффициента силуэта 0.7239 подтверждает, что кластеры расположены далеко друг от друга и являются достаточно плотными, что говорит о качественной кластеризации.

### 5. Повторение KMeans с 2 и 4 кластерами

На этом этапе необходимо повторить кластеризацию для двух и четырёх кластеров, чтобы сравнить результаты с оптимальным вариантом. Это позволяет увидеть, как изменение количества кластеров влияет на форму групп, качество разделения и значение коэффициента силуэта, а также убедиться, насколько корректно алгоритм отражает структуру данных при разных настройках.

```python
kmeans_2 = KMeans(n_clusters=2, random_state=42)
labels_2 = kmeans_2.fit_predict(df_scaled[['x_scaled', 'y_scaled']])

# Визуализация
centers_2 = kmeans_2.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(df_scaled['x_scaled'], df_scaled['y_scaled'],
            c=labels_2, cmap='viridis', s=40)
plt.scatter(centers_2[:, 0], centers_2[:, 1],
            c='red', s=200, marker='X')
plt.title("Кластеры (n_clusters = 2)")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

# Коэффициент силуэта
sil_2 = silhouette_score(df_scaled[['x_scaled', 'y_scaled']], labels_2)
print("Коэффициент силуэта (2 кластера):", sil_2)
```

```python
kmeans_4 = KMeans(n_clusters=4, random_state=42)
labels_4 = kmeans_4.fit_predict(df_scaled[['x_scaled', 'y_scaled']])

# Визуализация
centers_4 = kmeans_4.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(df_scaled['x_scaled'], df_scaled['y_scaled'],
            c=labels_4, cmap='viridis', s=40)
plt.scatter(centers_4[:, 0], centers_4[:, 1],
            c='red', s=200, marker='X')
plt.title("Кластеры (n_clusters = 4)")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

# Коэффициент силуэта
sil_4 = silhouette_score(df_scaled[['x_scaled', 'y_scaled']], labels_4)
print("Коэффициент силуэта (4 кластера):", sil_4)
```

При разбиении данных на два кластера алгоритм KMeans объединяет обе нижние группы в один общий кластер, а верхнюю группу выделяет отдельно. Такое решение не отражает истинной структуры данных, так как фактически в нижней части присутствуют две разные группы, различающиеся по оси x. Поэтому модель вынуждена объединять точки, которые лежат далеко друг от друга, что приводит к снижению качества кластеризации. Это подтверждается коэффициентом силуэта 0.497 — значением ниже оптимального, указывающим на среднюю степень разделённости кластеров и заметное ухудшение качества по сравнению с тремя кластерами.

При использовании четырёх кластеров нижние две группы остаются выделенными корректно, однако верхняя группа искусственно «разрезается» на два подклaстера. Такое деление не соответствует реальному распределению, так как в верхнем кластере нет внутреннего разрыва — точки образуют одну плотную область. Несмотря на это, коэффициент силуэта при четырёх кластерах равен 0.586, что выше, чем при двух кластерах, но ниже, чем при оптимальном разбиении на три группы. Это связано с тем, что KMeans создаёт более компактные подгруппы, однако они не отражают реальной структуры данных.

Таким образом, сравнение всех вариантов показывает, что наилучшим решением является кластеризация с тремя кластерами: она соответствует естественному распределению точек, обеспечивает чёткое разделение кластеров и демонстрирует максимальный коэффициент силуэта среди проверенных вариантов.

### 6. Построение графика по "методу локтя"

Метод локтя используется для выбора оптимального количества кластеров. На графике отображается зависимость инерции (суммы квадратов расстояний точек до своих центров) от числа кластеров. Инерция всегда уменьшается при увеличении числа кластеров, но после определённого момента улучшение становится минимальным. Точка, в которой происходит резкое замедление снижения кривой, и называется “локтем”. Она соответствует количеству кластеров, дающему лучшее соотношение качества разбиения и сложности модели.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled[['x_scaled', 'y_scaled']])
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia_values, marker='o', label='Inertia')

# Добавление "локтя"
plt.axvline(x=3, linestyle='--', color='red', label='Выбранное k = 3')

plt.title("Метод локтя")
plt.xlabel("Количество кластеров (k)")
plt.ylabel("Inertia")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

На графике зависимости инерции от количества кластеров отчётливо виден «локоть» в точке k = 3, где происходит резкое замедление снижения инерции. Это означает, что добавление дополнительных кластеров после значения 3 уже не приводит к существенному улучшению качества разбиения. Вертикальная пунктирная линия отмечает оптимальное число кластеров, которое обеспечивает наилучший баланс между точностью и сложностью модели.

На этапе предварительной обработки числовые признаки приводятся к единому масштабу с помощью стандартизации, что важно для корректной работы методов кластеризации. Категориальные признаки кодируются методом OneHotEncoder, а параметр drop='first' предотвращает появление избыточных столбцов и мультиколлинеарности. Такой подход позволяет объединить все преобразования в единую структуру, которая затем может применяться к данным перед кластеризацией.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Числовые признаки
numeric_features = [
    'age',
    'annual_income',
    'loan_amount',
    'interest_rate_pct',
    'credit_score',
    'num_open_accounts',
    'term_months'
]

# Категориальные признаки
categorical_features = [
    'employment_type',
    'education',
    'delinquency_30d',
    'residence_type'
]

# Препроцессор: стандартизация числовых + OHE для категориальных
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ]
)
```

Матрица диаграмм рассеяния строится по всем числовым признакам, чтобы визуально оценить распределения и взаимосвязи между переменными. Такое исследование помогает предварительно понять структуру данных и возможные закономерности перед применением методов кластеризации.

Некоторые особенности заметны и для признака interest_rate_pct: клиенты с более высокими процентными ставками несколько чаще относятся к группе дефолтов. При этом такие признаки, как возраст, количество открытых счетов и сумма кредита, не демонстрируют выраженного визуального разделения между классами. В целом диаграммы подтверждают наличие многомерного влияния сразу нескольких факторов, что делает последующую стандартизацию и кластеризацию важными этапами анализа.

Стандартизация числовых признаков выполняется для приведения всех показателей к единому масштабу: со средним значением 0 и стандартным отклонением 1. Это особенно важно для методов кластеризации, поскольку алгоритмы, основанные на расстояниях (например, k-means), чувствительны к различию в масштабе признаков. 

```python
from sklearn.preprocessing import StandardScaler

numeric_features = [
    'age',
    'annual_income',
    'loan_amount',
    'interest_rate_pct',
    'term_months',
    'credit_score',
    'num_open_accounts'
]

# Создание стандартизатора
scaler = StandardScaler()

# Масштабирование
X_numeric_scaled = scaler.fit_transform(df[numeric_features])

# Преобразуем в DataFrame для удобства анализа
df_scaled_num = pd.DataFrame(X_numeric_scaled, columns=numeric_features)

df_scaled_num.head()
```

На данном этапе необходимо выполнить кластеризацию клиентов методом k-means и определить оптимальное количество кластеров. Для этого применяется предварительно подготовленная матрица признаков, а затем оценивается качество разбиения при разных значениях k с помощью метода локтя и коэффициента силуэта. Эти подходы позволяют выбрать такое число кластеров, при котором структура данных отражается наиболее корректно, а группы объектов получаются максимально однородными и хорошо отделёнными друг от друга.

```python
# Применение препроцессора к данным
X_prepared = preprocessor.fit_transform(X)

# Посмотрим форму результирующей матрицы
X_prepared.shape

inertia_values = []
k_values = range(2, 10)

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_prepared)
    inertia_values.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_values, inertia_values, marker='o')
plt.title("Метод локтя для подбора количества кластеров")
plt.xlabel("Количество кластеров (k)")
plt.ylabel("Inertia")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

silhouette_scores = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X_prepared)
    score = silhouette_score(X_prepared, labels)
    silhouette_scores.append(score)

for k, s in zip(k_values, silhouette_scores):
    print(f"k = {k}: silhouette = {s}")

plt.figure(figsize=(8,5))
plt.plot(k_values, silhouette_scores, marker='o')
plt.title("Коэффициент силуэта для разных значений k")
plt.xlabel("Количество кластеров (k)")
plt.ylabel("Silhouette Score")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

График метода локтя показывает плавное снижение инерции при увеличении количества кластеров. Наибольшее уменьшение наблюдается при переходе от 2 к 3 кластерам, а затем — от 3 к 4. После k = 4 снижение становится значительно более пологим, без выраженного резкого излома. Это указывает на слабовыраженную структуру кластеров в данных и отсутствие чётко выделяемых групп.

График коэффициента силуэта подтверждает этот вывод: значения силуэта низкие для всех k (порядка 0.08–0.096), что говорит о слабой отделимости кластеров. Максимальные значения наблюдаются при k = 2 (0.0968) и k = 3 (0.0957), а затем метрика начинает уменьшаться. Наиболее заметное падение происходит при k = 4, что подтверждает неоптимальность разбиения на четыре и более кластера.

С учётом обоих методов оптимальным является выбор k = 2 или k = 3, где достигается наибольшее значение силуэта и наиболее выраженное изменение инерции. В практическом контексте предпочтительным считается k = 2, так как оно даёт максимально высокую метрику силуэта и предлагает более простую, интерпретируемую структуру данных.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

k_opt = 2

# Обучение модели
kmeans_vis = KMeans(n_clusters=k_opt, random_state=42)
labels_vis = kmeans_vis.fit_predict(X_prepared)

# PCA для визуализации
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_prepared)

plt.figure(figsize=(8, 6))

# Цвета для кластеров
colors = plt.cm.viridis(np.linspace(0, 1, k_opt))

# Отрисовка каждого кластера отдельно, чтобы легенда работала корректно
for cluster_id in range(k_opt):
    mask = labels_vis == cluster_id
    plt.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        s=30,
        alpha=0.7,
        color=colors[cluster_id],
        label=f'Кластер {cluster_id}'
    )

# Центры в PCA-пространстве
centers_pca = pca.transform(kmeans_vis.cluster_centers_)
plt.scatter(
    centers_pca[:, 0],
    centers_pca[:, 1],
    c='red',
    s=200,
    marker='X',
    label='Центры кластеров'
)

plt.title(f"Визуализация кластеров после KMeans (k = {k_opt})")
plt.xlabel("PCA компонент 1")
plt.ylabel("PCA компонент 2")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

График показывает распределение объектов в пространстве первых двух главных компонент PCA после применения алгоритма KMeans с двумя кластерами. Две группы визуально частично разделяются вдоль первой компоненты: левая часть пространства преимущественно занята объектами кластера 0, а правая — объектами кластера 1. При этом значительная часть точек располагается близко друг к другу и образует плотную центральную область, что указывает на слабое визуальное разделение кластеров и наличие пересечений между ними. Центры кластеров расположены недалеко друг от друга и занимают центральные позиции внутри соответствующих групп точек, что является ожидаемым поведением для алгоритма KMeans. В целом график демонстрирует наличие двух сгущений, однако границы между ними выражены недостаточно чётко, что отражает сложную структуру данных и слабую природную сегментацию клиентов.

После выполнения кластеризации выбранное количество кластеров позволяет разделить заемщиков на группы с различными финансовыми характеристиками. Анализ средних значений признаков внутри каждого кластера помогает выявить особенности каждой группы. Один из кластеров может характеризоваться более низким доходом, меньшим кредитным скором и повышенной процентной ставкой, что делает его более рискованным с точки зрения дефолта. Другой кластер, напротив, может объединять клиентов с более высоким уровнем дохода и лучшей кредитной историей. Такое разбиение показывает, какие потребительские сегменты присутствуют в данных, и позволяет интерпретировать ключевые факторы, влияющие на вероятность дефолта.

```python
k_opt = 2

kmeans_final = KMeans(n_clusters=k_opt, random_state=42)
labels_final = kmeans_final.fit_predict(X_prepared)

df['cluster'] = labels_final

cluster_means_numeric = df.groupby('cluster').mean(numeric_only=True)
print("Средние числовые признаки по кластерам:")
print(cluster_means_numeric)
```

Кластеры демонстрируют явное различие по ключевым финансовым характеристикам. Кластер 0 представляет собой группу более надёжных заемщиков: у них выше средний годовой доход, меньше сумма кредита, ниже процентная ставка и значительно выше кредитный скор. Эта группа показывает минимальный уровень дефолта — всего около 5.6%. Средний возраст и количество открытых счетов также находятся на умеренном уровне.

Кластер 1, напротив, включает клиентов с более высоким кредитным риском. Их средний кредитный скор заметно ниже, сумма кредита почти в два раза выше, а процентная ставка — значительно выше, что отражает более жёсткие условия кредита. Уровень дефолтов в этой группе достигает 42%, что указывает на её проблемность. Средний доход в этом кластере ниже, а возраст немного выше, чем в первом кластере.

Таким образом, первый кластер характеризует финансово стабильных заемщиков с низким риском, тогда как второй кластер объединяет более рискованных клиентов с ухудшенными кредитными показателями и повышенной вероятностью дефолта.

Silhouette Score показывает степень разделённости кластеров: чем ближе значение к 1, тем лучше сформированы кластеры.
Индекс Calinski–Harabasz отражает качество разброса объектов внутри кластеров относительно дистанции между кластерами: большие значения означают более чёткую структуру.
Показатель Davies–Bouldin измеряет сходство между кластерами: чем меньше значение, тем лучше они разделены.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# labels_final — итоговые метки кластеров
# X_prepared — данные после препроцессинга

silhouette = silhouette_score(X_prepared, labels_final)
calinski = calinski_harabasz_score(X_prepared, labels_final)
davies = davies_bouldin_score(X_prepared, labels_final)

print("Silhouette Score:", silhouette)
print("Calinski–Harabasz Index:", calinski)
print("Davies–Bouldin Index:", davies)
```

Полученные значения метрик указывают на то, что структура кластеров выражена слабо. Коэффициент силуэта 0.0968 находится близко к нулю, что свидетельствует о значительном пересечении групп и небольшой разнице между расстояниями внутри и между кластерами. Это означает, что объекты разных кластеров расположены близко друг к другу и не формируют отчётливо разделённых областей.

Индекс Calinski–Harabasz (308) отражает умеренное соотношение межкластерной и внутрикластерной дисперсии: разброс объектов внутри кластеров остаётся довольно высоким, а различия между центрами кластеров — незначительными. Такое значение также подтверждает слабую кластерную структуру.

Значение индекса Davies–Bouldin (3.00) достаточно велико, что указывает на значительное сходство между кластерами: группы находятся близко и обладают большой внутренней вариативностью. Для хорошо разделённых кластеров этот показатель значительно ниже.

В целом, метрики показывают, что выбранное разбиение не выявляет ярко выраженных естественных групп в данных, а обнаруженная сегментация отражает только слабые различия между объектами.

```python
from yellowbrick.cluster import SilhouetteVisualizer, KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X = df_scaled[['x_scaled', 'y_scaled']].values

ks = range(2, 5)
silhouette_scores = []

for k in ks:
    model = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = model.fit_predict(X)

    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

    fig, ax = plt.subplots(figsize=(6, 4))
    viz = SilhouetteVisualizer(model, ax=ax)
    viz.fit(X)

    ax.set_title(f"Силуэтная диаграмма для k = {k}")
    ax.set_xlabel("Значение коэффициента силуэта")
    ax.set_ylabel("Кластеры")

    viz.show()

model = KMeans(random_state=42, n_init="auto")

viz = KElbowVisualizer(
    model,
    k=(2, 8),
    metric="silhouette",
    timings=False
)

viz.fit(X)

viz.ax.set_title("Выбор оптимального числа кластеров по коэффициенту силуэта")
viz.ax.set_xlabel("Количество кластеров (k)")
viz.ax.set_ylabel("Средний коэффициент силуэта")

viz.show()

viz.elbow_value_
```

По силуэтным диаграммам видно, что при k = 3 кластеры наиболее чётко разделены: средний коэффициент силуэта максимален, почти все значения положительные и достаточно высокие, а сами кластеры имеют сопоставимый размер без выраженных проблемных областей. При k = 2 часть данных оказывается объединённой в слишком крупный кластер, из-за чего качество разделения ниже. При k = 4 появляется перекластеризация: некоторые кластеры становятся менее устойчивыми, а средний silhouette уменьшается.

Это подтверждается и графиком зависимости silhouette от числа кластеров: наблюдается явный максимум при k = 3 (около 0.72), после чего качество кластеризации снижается. Таким образом, оптимальным числом кластеров для данных является 3, что согласуется как с визуальным анализом силуэтных диаграмм, так и с результатом метода «локтя» по коэффициенту силуэта.

Задание 6. Выполнить подсчёт 4 любых метрик для разного количества кластеров (от 2 до 10). Построить график - по оси х - кол-во кластеров, по у - значение метрики (отразить все 4 метрики на одном графике разным цветом).

В рамках задания были рассчитаны четыре метрики качества кластеризации (Silhouette score, индекс Калински–Харабаза, индекс Дэвиса–Болдина и внутрикластерная сумма квадратов Inertia) для различного числа кластеров от 2 до 10 с использованием алгоритма k-средних. Для каждой метрики были получены значения при изменении количества кластеров, после чего все результаты были отображены на одном графике для наглядного сравнения их поведения.

```python
from sklearn.metrics import (
    calinski_harabasz_score,
    davies_bouldin_score
)

X = df_scaled[['x_scaled', 'y_scaled']].values

ks = range(2, 11)

silhouette_scores = []
calinski_scores = []
davies_scores = []
inertia_scores = []

for k in ks:
    model = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = model.fit_predict(X)

    silhouette_scores.append(silhouette_score(X, labels))
    calinski_scores.append(calinski_harabasz_score(X, labels))
    davies_scores.append(davies_bouldin_score(X, labels))
    inertia_scores.append(model.inertia_)

# Построение графика
plt.figure(figsize=(8, 5))

plt.plot(ks, silhouette_scores, marker='o', label='Silhouette score')
plt.plot(ks, calinski_scores, marker='o', label='Calinski-Harabasz index')
plt.plot(ks, davies_scores, marker='o', label='Davies-Bouldin index')
plt.plot(ks, inertia_scores, marker='o', label='Inertia (WCSS)')

plt.xlabel("Количество кластеров (k)")
plt.ylabel("Значение метрики")
plt.title("Зависимость метрик качества кластеризации от числа кластеров")
plt.legend()
plt.grid(True)

plt.show()
```

По графику видно, что метрики Silhouette score и Calinski–Harabasz index достигают максимальных значений при k = 3, что указывает на наилучшее качество кластеризации при данном числе кластеров. Индекс Davies–Bouldin принимает минимальное значение также в области k = 3, что подтверждает хорошую разделимость кластеров. Метрика Inertia монотонно убывает с ростом числа кластеров, однако наиболее резкое снижение наблюдается до k = 3, после чего улучшения становятся менее значимыми. Таким образом, все рассмотренные метрики согласованно указывают на k = 3 как оптимальное число кластеров для данных.

Задание 7. Выполнить нормализацию данных с помощью MinMaxScaler, MaxAbsScaler и сравнить полученные итоговые метрики с теми, что были получены на данных, нормализованных с помощью StandardScaler.

В ходе задания были рассмотрены различные способы нормализации данных: StandardScaler, MinMaxScaler и MaxAbsScaler. Для каждого метода масштабирования данные преобразовывались отдельно, после чего выполнялась кластеризация алгоритмом k-средних при фиксированном числе кластеров. Для оценки качества кластеризации были рассчитаны несколько метрик, значения которых затем были сведены в таблицу для наглядного сравнения влияния метода нормализации на результат.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler

X = df_scaled[['x_scaled', 'y_scaled']].values

scalers = {
    "StandardScaler": StandardScaler(),
    "MinMaxScaler": MinMaxScaler(),
    "MaxAbsScaler": MaxAbsScaler()
}

scaled_data = {}

for name, scaler in scalers.items():
    scaled_data[name] = scaler.fit_transform(X)
k = 3

results = {}

for name, X_scaled in scaled_data.items():
    model = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = model.fit_predict(X_scaled)

    results[name] = {
        "Silhouette": silhouette_score(X_scaled, labels),
        "Calinski-Harabasz": calinski_harabasz_score(X_scaled, labels),
        "Davies-Bouldin": davies_bouldin_score(X_scaled, labels),
        "Inertia": model.inertia_
    }

results_df = pd.DataFrame(results).T
results_df
```

По полученным значениям метрик видно, что наилучшее качество кластеризации достигается при использовании StandardScaler: коэффициент силуэта и индекс Калински–Харабаза принимают максимальные значения, а индекс Дэвиса–Болдина — минимальное. Методы MinMaxScaler и MaxAbsScaler дают сопоставимые, но несколько худшие результаты, что связано с иным масштабированием расстояний между объектами. Таким образом, для рассматриваемых данных и алгоритма k-средних использование StandardScaler является наиболее предпочтительным способом нормализации.

Задание 9. Выполнить кластеризацию с помощью метода Affinity Propagation. Добавить описание метода (теор. часть). Сравнить результаты с полученными ранее.

В рамках задания была выполнена кластеризация данных с использованием метода Affinity Propagation. Алгоритм применялся к стандартизированным данным без предварительного задания числа кластеров, при этом количество кластеров определялось автоматически на основе структуры данных. Для оценки качества полученного разбиения были рассчитаны стандартные метрики кластеризации: коэффициент силуэта, индекс Калински–Харабаза и индекс Дэвиса–Болдина. Результаты были представлены в виде таблицы.

```python
from sklearn.cluster import AffinityPropagation


X = df_scaled[['x_scaled', 'y_scaled']].values

ap = AffinityPropagation(random_state=42)
labels_ap = ap.fit_predict(X)

n_clusters_ap = len(np.unique(labels_ap))

metrics_ap = {
    "Количество кластеров": n_clusters_ap,
    "Silhouette": silhouette_score(X, labels_ap),
    "Calinski-Harabasz": calinski_harabasz_score(X, labels_ap),
    "Davies-Bouldin": davies_bouldin_score(X, labels_ap)
}

ap_results_df = pd.DataFrame.from_dict(
    metrics_ap, orient="index", columns=["Значение"]
)

ap_results_df
```

Метод Affinity Propagation автоматически выделил 7 кластеров, однако значения метрик качества указывают на менее выраженную кластерную структуру по сравнению с результатами, полученными ранее методом k-средних. Низкое значение коэффициента силуэта и повышенное значение индекса Дэвиса–Болдина свидетельствуют о слабой разделимости и частичном перекрытии кластеров. Таким образом, для данных более подходящим оказался метод k-средних с предварительным выбором числа кластеров, тогда как Affinity Propagation в данном случае привёл к переоценке количества кластеров и снижению качества кластеризации.
